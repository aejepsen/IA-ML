Importar Base de Dados

```python
#abril2018_df = pd.read_csv(r'dataset/abril2018.csv')
#display(abril2018_df)

caminho_bases = pathlib.Path('/content/drive/MyDrive/#treinamentos Python/analise-dados-panda/Projeto 3 - Ciências de Dados/data_science/dataset')
# cria dataframe vazio
base_airbnb = pd.DataFrame()
meses = {'jan':1, 'fev':2, 'mar':3, 'abr':4, 'mai':5, 'jun':6, 'jul':7, 'ago':8, 'set':9, 'out':10, 'nov':11, 'dez':12}
for arquivo in caminho_bases.iterdir():
    # incluir columa mes
    # abril2008.csv
    nome_mes = arquivo.name[:3]
    # inclui valor correspondente a cada mes
    mes = meses[nome_mes]
    # incluir columa ano
    # abril2008.csv
    ano = arquivo.name[-8:]
    ano = int(ano.replace('.csv', ''))
    # concatena caminho com arquivo
    df = pd.read_csv(caminho_bases / arquivo.name)
    df['ano'] = ano
    df['mes'] = mes
    # inclui arquivo do dataframe
    base_airbnb = base_airbnb.append(df)
    
display(base_airbnb)
```
